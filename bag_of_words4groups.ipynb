```python
import pandas as pd
import age_grouping as ag
import keywords_by_group as kbg
import warnings
warnings.filterwarnings("ignore")
```

```python
EPData = pd.read_csv("../First100000.csv")
```

```python
## Group data by a given bin size
age_range = [13,78]
bin_size = 11
Data = ag.age_filtering(EPData,age_range)
Data = ag.age_grouping(Data,bin_size,age_range)
group = list(set(Data['age_group']))
nfeature = 50
```

```python
for group_var in ['age_group','gender']:
    for text_var in ['content']:
        if group_var == 'age_group':
            kbg.keyword_by_group(Data,group_var,text_var,nfeature,False,True)
        else:
            kbg.keyword_by_group(Data.dropna(),group_var,text_var,nfeature,False,True)
```
